```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

plt.rcParams['figure.figsize'] = (12.0 ,9.0)
```

```python
data = pd.read_csv('data.csv')
x = data.iloc[0:24,0]
y = data.iloc[0:24, 1]
plt.scatter (x,y)
plt.show()
```

```python
#Building the model

w = 0
b = 0
l = 0.0001    # l is inversly prop to epochs
epobhs = 1000

n = float (len(x))

for i in range (len(x)):
  y_pred = w*x + b
  dw = (-1/n) * sum(x * (y - y_pred))
  db = (-1/n) * sum(y - y_pred)
  w = w - l*dw
  b = b - l*db

print(w )
print(b)
```

```python
# making predictions

y_pred = w*x + b
plt.scatter(x,y)
plt.plot (x,y_pred, color = 'orange', marker = '*')
plt.show()
```
