# Linear Regression using random numbers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# num of vales
n = random.randint(0,9)
x = np.empty(n)
for i in range (n):
  a = random.randint(0,100)
  x[i] = a
print("x= ", x)
x = x.reshape(-1,1)

y = np.empty(n)

for i in range (n):
  a = random.randint(0,100)
  y[i] = a
print("y= " ,y)
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR = LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred = SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,pred)
```

```python
plt.plot(x,pred , color='Red', marker = '*')
print("MSE: ", mse)
print("Intercept: ", SLR.intercept_)
print("Slope:" , SLR.coef_)
```

# Linear Regression using Exel data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
x = df.loc[0:12 , ('x')]
y = df.loc[0:12 , ('y')]

x = x.to_numpy()
y = y.to_numpy()

print('x= ', x)
print('y= ',y)
```

```python
x = x.reshape((-1,1))
```

```python
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR = LinearRegression()
SLR.fit(x,y)
```

```python
pred = SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y , pred)

plt.scatter(x,y, label = 'origrinal')
plt.plot(x,pred, color = 'black', marker = '*')
plt.legend()
print("MSE= ", mse)
print ("Y intercept= ", SLR.intercept_)
print("Slope = ", SLR.coef_)
```
